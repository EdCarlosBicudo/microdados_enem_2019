O Enem conta com cinco notas em diferentes matérias: Ciências da Natureza, Ciências Humanas, Linguagens e Códigos, Matemática e Redação.

Essas notas podem ser utilizadas de maneiras diferentes e com pesos diferentes por cada instituição de ensino superior.

Aqui será utlizado como nota final a média simples dessas cinco notas.

```python
#Subistituindo notas faltantes por 0
dados['NOTA CIENCIAS DA NATUREZA'] = dados['NOTA CIENCIAS DA NATUREZA'].fillna(0)
dados['NOTA CIENCIAS HUMANAS'] = dados['NOTA CIENCIAS HUMANAS'].fillna(0)
dados['NOTA LINGUAGENS E CODIGOS'] = dados['NOTA LINGUAGENS E CODIGOS'].fillna(0)
dados['NOTA MATEMATICA'] = dados['NOTA MATEMATICA'].fillna(0)
dados['NOTA REDACAO'] = dados['NOTA REDACAO'].fillna(0)
```

```python
dados['MEDIA FINAL'] = ((dados['NOTA CIENCIAS DA NATUREZA'] + dados['NOTA CIENCIAS HUMANAS'] + 
                         dados['NOTA LINGUAGENS E CODIGOS'] + dados['NOTA MATEMATICA'] + 
                         dados['NOTA REDACAO'])/5)
```
